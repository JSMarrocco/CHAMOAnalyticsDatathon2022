# Simple NN

```python
df = utils.load_and_clean_master_dataset()

df_train, df_val, df_test = utils.split_train_test_val(df)

X_train, y_train = utils.split_Xy(df_train)
X_val, y_val = utils.split_Xy(df_val)
X_test, y_test = utils.split_Xy(df_test)

i = 4
X_train, X_val, X_test = X_train[utils.cols[:i+1]], X_val[utils.cols[:i+1]], X_test[utils.cols[:i+1]]

# Take into account sliding window of models
X_test, y_test = X_test[20:], y_test[20:]
```

# Transformer

**Prepare data**

```python
df = utils.load_and_clean_master_dataset()

df_train, df_val, df_test = utils.split_train_test_val(df)

X_train, y_train = utils.split_Xy(df_train)
X_val, y_val = utils.split_Xy(df_val)
X_test, y_test = utils.split_Xy(df_test)

i = 4
X_train, X_val, X_test = X_train[utils.cols[:i+1]], X_val[utils.cols[:i+1]], X_test[utils.cols[:i+1]]

X_train, y_train = utils.to_batch(X_train, y_train)
X_val, y_val = utils.to_batch(X_val, y_val)
X_test, y_test = utils.to_batch(X_test, y_test)
```

**Reload model**

```python
crnt_dir = "..\modeling" # crnt_dir = pathlib.Path('.').parent.resolve()
transformer_models = []
for i in range(2):
    transformer_models.append(keras.models.load_model(crnt_dir+'\\transformer'+f'\\model_{i+1}'))
transformer_model = keras.models.load_model(crnt_dir+'\\transformer'+'\model_1')
```

```python
for layer in transformer_models[0].layers:
        print("Layer Name: " + layer.name)
        print("Type of layer: " + layer.__class__.__name__)
        print("Input dimesion: {}".format(layer.input_shape[1:]))
        print("Output dimesion: {}".format(layer.output_shape[1:]))
        print("Parameter Count: {}".format( layer.count_params()))
        try:
            print("Activation : " + layer.activation.__name__)
            print(" ")
        except:
            print(" ")
```

**Evaluate model**

```python

y_pred_transformers = []
for model in transformer_models[1:]:
    y_pred_transformers.append(model(X_test))

y_pred_transformer_best = transformer_model(X_test)
mean_squared_error(y_test.to_numpy(), y_pred_transformer_best.numpy().reshape(-1))
```

# Ensemble

```python
y_pred_nns = [y_pred_nn.reshape((-1,1)) for y_pred_nn in y_pred_nns ]
y_pred_ensemble = np.mean([*y_pred_nns, *y_pred_lstms, *y_pred_transformers], axis=0)
# y_pred_ensemble = np.mean([*y_pred_nns, *y_pred_transformers], axis=0)
ensemble_mse = mean_squared_error(y_test.to_numpy(), y_pred_ensemble.reshape(-1))
```

```python
np.array(y_pred_nns).shape
```

```python
print(f'Final (ensemble) MSE is {ensemble_mse}')
```

```python
pmin = y_pred_ensemble.min(axis=0)
pmax = y_pred_ensemble.max(axis=0)

drawdown = (pmin-pmax) / pmax 
print(f"Mean Drawdown: {np.mean(drawdown)}")

bias = np.abs(np.mean(y_pred_ensemble - np.expand_dims(y_test, axis=-1)))
print(f"Mean Bias: {np.mean(bias)}")

variance = np.mean(np.var([*y_pred_nns, *y_pred_lstms, *y_pred_transformers], axis=0))
# variance = np.mean(np.var([*y_pred_nns, *y_pred_transformers], axis=0))
print(f"variance: {variance}")
```
